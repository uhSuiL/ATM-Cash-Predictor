```python
import matplotlib.pyplot as plt

from model.util import visualize_train
import pandas as pd
import os
```

```python
log_dir = '../log/ClusteredSimpleGRU/'
param = '(norm, train_h0=True, win=10, batch_size=4)'
time = '2024-05-27-20-48'

log_data_list = []
for f in os.listdir(log_dir):
    print(os.path.join(log_dir, f, param, time))
    
    _ = visualize_train(
        os.path.join(log_dir, f, param, time), 
        ['test_loss', 'MAE'],
        v_lines = [50, 80]
    )
    log_data_list.append(_)
```

```python
log_dir = '../log/ClusteredStrategicGRU/'
param = '(norm, train_h0=True, win=10, batch_size=4)'
time = '2024-05-27-22-43'

log_data_list2 = []
for f in os.listdir(log_dir):
    print(os.path.join(log_dir, f, param, time))
    
    _ = visualize_train(
        os.path.join(log_dir, f, param, time), 
        ['test_loss', 'MAE'],
        v_lines = [50, 80]
    )
    log_data_list2.append(_)
```

```python
log_dir = '../log/ClusteredStrategic(MLP)GRU/'
param = '(norm, train_h0=True, win=10, batch_size=4)'
time = '2024-05-27-22-54'

log_data_list3 = []
for f in os.listdir(log_dir):
    print(os.path.join(log_dir, f, param, time))
    
    _ = visualize_train(
        os.path.join(log_dir, f, param, time), 
        ['test_loss', 'MAE'],
        v_lines = [50, 80]
    )
    log_data_list3.append(_)
```

```python
for i in range(len(log_data_list)):
    log_data_list[i]['test_loss'].rename(f'simple_cluster_{i}_test_loss').plot()
    log_data_list2[i]['test_loss'].rename(f'strategic_cluster_{i}_test_loss').plot()
    log_data_list3[i]['test_loss'].rename(f'strategic(mlp)_cluster_{i}_test_loss').plot()
    plt.legend()
    plt.show()
```
